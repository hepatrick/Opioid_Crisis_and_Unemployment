```python
# Import dependencies
 import pandas as pd
```

```python
# Path to CSV file

file = "Resources/State_Unemployment_Rates_2014.csv"

# Read data in to Pandas

unemp_df = pd.read_csv(file)
unemp_df.head()
```

```python
# Set state name as the index 

unemp_df2 = unemp_df.set_index('State')

unemp_df2.head()
```

```python
# Drop rows that do not exist in the other dataset and NaN
# DC & Puerto Rico

updated_df = unemp_df.drop(unemp_df.index[[8, 39, 52]])
updated_df.head()
```

```python
# Check data types
updated_df.dtypes
```

```python
# Convert October to a float - it is currently an object

updated_df["October"] = updated_df.October.astype(float)
```

```python
# Check data types again
updated_df.dtypes
```

```python
updated_df.columns
```

```python
# Create new column for the average 
# updated_df['Avg_Unemployment'] = (updated_df.January + updated_df.February + updated_df.March + updated_df.April + updated_df.May + updated_df.June + updated_df.July + updated_df.August + updated_df.September + updated_df.October + updated_df["November "] + updated_df.December)/12
```

```python
updated_df['Avg_Unemployment'] = updated_df[[col for col in updated_df.columns if 'State' not in col]].sum(axis=1) / 12
```

```python
updated_df.head()
```

```python
updated_df.to_csv(r'C:updated_unemployment2.csv')
```

```python
# Merge unemp data with opioid death rate data 

# Path to CSV file

opioid_file = "Resources/overdoses_clean.csv"

# Read data in to Pandas

op_df = pd.read_csv(opioid_file)
op_df.head()
```

```python
merged = pd.merge(op_df, updated_df, on="State")
merged.head()
```

```python
final_df = merged.drop(columns=['Unnamed: 0', 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November ', 'December'])
final_df.head()
```

```python
final_df.to_csv(r'C:merged_unemployment_opioid.csv')
```
